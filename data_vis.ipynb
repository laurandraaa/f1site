```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
places = {'drivers': ['perez', 'alonso', 'sainz', 'russell', 'ocon', 'stroll'], 
          'prediction': [4.62, 5.83, 6.91, 7.01, 8.28, 9.13]}
places_df = pd.DataFrame(data = places)

places_df
```

```python
pred_list = places_df['prediction'].values.tolist()
pred_list
```

```python
def vis_places(df):
    fig, ax1 = plt.subplots()
    for ind in df.index:
        x = np.arange(4, 10, 0.001)
        print(df['prediction'][ind])
        plt.plot(x, norm.pdf(x, float(df['prediction'][ind]), .5))
```

```python
vis_places(places_df)
```
